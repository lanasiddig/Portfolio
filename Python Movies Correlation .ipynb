```python
#Scatter plot with budget vs gross
plt.scatter(x=df['Budget'],y=df['Gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget of Film')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
#lets start looking at the correlation (works only in numerical fields)
#pearson, kendall, spearman correlation 
df.corr(method='pearson') #pearson aka default one
```

```python
df.corr(method='kendall') #kendall correlation
```

```python
df.corr(method='spearman') #spearman correlation 
```

```python
#high correlation between budget and gross is true 
```

```python
correlation_matrix=df.corr(method='pearson')
sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlatin Matrix for Numerical Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
#comapny correlation (non-numerical to numerical values)
df_numerized=df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype=='object'):
        df_numerized[col_name]=df_numerized[col_name].astype('category')
        df_numerized[col_name]=df_numerized[col_name].cat.codes
df_numerized  
df_numerized.head() 
```

```python
df.head()
```

```python
correlation_matrix=df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlatin Matrix for Numerical Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized.corr()
```

```python
#unstaking for the purpose of obtaining the highest correlation 
correlation_mat=df_numerized.corr()
corr_pairs=correlation_mat.unstack()
corr_pairs
corr_pairs.head()
```

```python
sorted_pairs=corr_pairs.sort_values()
sorted_pairs
sorted_pairs.head()
```

```python
high_correlation=sorted_pairs[(sorted_pairs)>0.5]
high_correlation
```
